```python
import pandas as pd
df = pd.read_csv('power.csv')
def baltic(country):
    if country in ['Estonia','Latvia','Lithuania']:
        return 'Baltica'
    else:
        return 'Other'
df['baltica'] = df.country.apply(baltic)
df_filtered = df[(df.baltica == 'Baltica') & (df['year'] >= 2005) & (df['year'] <= 2010) & ((df['category'] == 4) | (df['category'] == 21)| (df['category'] == 12)) & (df['quantity'] > 0)]
summ = df_filtered.quantity.sum()
print('Страны Прибалтики потребляют суммарно =', summ)
```
