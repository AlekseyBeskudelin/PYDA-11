1. Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.


```python
import pandas as pd
df = pd.read_csv('ml-latest-small/ratings.csv')
max_ = df.groupby('userId').max()[['timestamp']].reset_index()
max_.rename(columns={'timestamp': 'timestampmax'}, inplace=True)
min_ = df.groupby('userId').min()[['timestamp']].reset_index()
min_.rename(columns={'timestamp': 'timestampmin'}, inplace=True)
times = df.groupby('userId').count()[['rating']].reset_index()
times.rename(columns={'rating': 'time'}, inplace=True)
times_max = times.loc[(times.time >= 100)]
with_lifetime = times_max.merge(min_, how='left').merge(max_)
with_lifetime['lifetime'] = with_lifetime['timestampmax'] - with_lifetime['timestampmin']
with_lifetime

```
